```python
import numpy as np
import os
import sys

# Set up paths to find the modules
ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
AGENTS_DIR = os.path.join(ROOT, "agents")

if AGENTS_DIR not in sys.path:
    sys.path.insert(0, AGENTS_DIR)
if ROOT not in sys.path:
    sys.path.insert(0, ROOT)

from run_simulation import run_simulation
from plots import plot_realized_vs_expected, plot_expected_components

# Run the simulation
sender, receiver, game = run_simulation(
    T=30000,
    n=2,
    p_observe_state=0.2,
    seed=10,
    record=True
)

# Print final statistics
print(f"Final success rate: {np.mean(game.success_history):.4f}")
print(f"Final expected payoff (mixed): {game.expected_payoff_mixed[-1]:.4f}")
print(f"Final expected payoff (signal only): {game.expected_payoff_signal_only[-1]:.4f}")
print(f"Final expected payoff (state+signal): {game.expected_payoff_state_signal[-1]:.4f}")

# Plot results
plot_realized_vs_expected(game)
plot_expected_components(game)
```
